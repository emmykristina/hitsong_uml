```python
# =========================
# CELL 1: IMPORTS
# =========================
# pandas = läsa och hantera tabeller
# numpy = hantera NaN och numeriska operationer
# pathlib = bygga robusta filvägar
# matplotlib/seaborn = visualisering

import pandas as pd
import numpy as np
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# =========================
# CELL 2: LOAD RAW DATA
# =========================
# Vi hittar projektmappen oavsett om notebooken körs från /notebooks eller projektroten.

project_folder = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

# OBS: ändra filnamnet här om csv-filen heter något annat i er data-mapp
raw_path = project_folder / "data" / "spotify.csv"

print("Projektmapp:", project_folder)
print("Datafil:", raw_path)
print("Finns filen?", raw_path.exists())

df_raw = pd.read_csv(raw_path)

df_raw.head()
```

```python
# =========================
# CELL 3: FIRST OVERVIEW
# =========================
# Här kollar vi storlek, kolumner och datatyper.

print("Shape:", df_raw.shape)

print("\nKolumner:")
print(df_raw.columns.tolist())

print("\nInfo:")
df_raw.info()
```

```python
# =========================
# CELL 4: MISSING VALUES
# =========================
# Här ser vi vilka kolumner som har saknade värden.

missing = df_raw.isna().sum().sort_values(ascending=False)

missing[missing > 0]
```

```python
# =========================
# CELL 5: DUPLICATES
# =========================
# Fullständiga dubbletter = hela raden är identisk.
# Track_id-dubbletter = samma låt förekommer flera gånger, ofta i olika genrer.

print("Full row duplicates:", df_raw.duplicated().sum())
print("Duplicate track_id:", df_raw["track_id"].duplicated().sum())
print("Unique track_id:", df_raw["track_id"].nunique())
```

```python
# =========================
# CELL 6: EXAMPLE DUPLICATE TRACK IDS
# =========================
# Vi tittar på några låtar som förekommer flera gånger.

df_raw[df_raw.duplicated(subset=["track_id"], keep=False)] \
    .sort_values("track_id") \
    [["track_id", "artists", "track_name", "track_genre", "popularity"]] \
    .head(20)
```

```python
# =========================
# CELL 7: NUMERIC RANGES
# =========================
# Här kollar vi om numeriska variabler ligger inom rimliga gränser.

numeric_cols = df_raw.select_dtypes(include="number").columns

df_raw[numeric_cols].describe().T
```

```python
# =========================
# CELL 8: SPECIAL VALUE CHECKS
# =========================
# Några värden är tekniskt giltiga i datan men kan vara problematiska:
# - duration_ms = 0 betyder låten har 0 längd, troligen fel
# - tempo = 0 kan betyda att tempo inte kunde beräknas
# - time_signature = 0 är svårt att tolka musikaliskt

print("duration_ms = 0:", (df_raw["duration_ms"] == 0).sum())
print("tempo = 0:", (df_raw["tempo"] == 0).sum())
print("time_signature = 0:", (df_raw["time_signature"] == 0).sum())

bounded_cols = [
    "danceability",
    "energy",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence"
]

for col in bounded_cols:
    invalid = ((df_raw[col] < 0) | (df_raw[col] > 1)).sum()
    print(f"{col} outside 0-1:", invalid)
```

```python
# =========================
# CELL 9: CREATE CLEANING COPY
# =========================
# Vi skapar en kopia av rådata.
# På så sätt ändrar vi inte originaldatan direkt.

df_clean_base = df_raw.copy()

print("Original shape:", df_raw.shape)
print("Cleaning copy shape:", df_clean_base.shape)
```

```python
# =========================
# CELL 10: REMOVE UNNECESSARY INDEX COLUMN
# =========================
# 'Unnamed: 0' är en gammal indexkolumn från csv-exporten.
# Den säger inget om musiken och behövs inte i analysen.

if "Unnamed: 0" in df_clean_base.columns:
    df_clean_base = df_clean_base.drop(columns=["Unnamed: 0"])

print("Columns after removing Unnamed: 0:")
print(df_clean_base.columns.tolist())
```

```python
# =========================
# CELL 11: REMOVE MISSING CORE TEXT INFO
# =========================
# För vår story behöver vi veta artist, album och låtnamn.
# Eftersom det bara verkar vara 1 rad som saknar detta tar vi bort den.

before = df_clean_base.shape[0]

df_clean_base = df_clean_base.dropna(
    subset=["artists", "album_name", "track_name"]
).copy()

after = df_clean_base.shape[0]

print("Rows before:", before)
print("Rows after:", after)
print("Removed rows:", before - after)
```

```python
# =========================
# CELL 12: REMOVE IMPOSSIBLE DURATION
# =========================
# En låt med duration_ms = 0 betyder att låten är 0 millisekunder lång.
# Det är inte rimligt för vår analys, så vi tar bort sådana rader.

before = df_clean_base.shape[0]

df_clean_base = df_clean_base[df_clean_base["duration_ms"] > 0].copy()

after = df_clean_base.shape[0]

print("Rows before:", before)
print("Rows after:", after)
print("Removed rows:", before - after)
```

```python
# =========================
# CELL 13: HANDLE TEMPO AND TIME SIGNATURE ZERO VALUES
# =========================
# tempo = 0 och time_signature = 0 kan betyda att analysvärdet saknas eller är svårt att tolka.
# Vi tar inte bort hela låten.
# Vi sätter dessa specifika värden till NaN så att de kan hanteras senare.

df_clean_base.loc[df_clean_base["tempo"] == 0, "tempo"] = np.nan
df_clean_base.loc[df_clean_base["time_signature"] == 0, "time_signature"] = np.nan

print("tempo missing after cleaning:", df_clean_base["tempo"].isna().sum())
print("time_signature missing after cleaning:", df_clean_base["time_signature"].isna().sum())
```

```python
# =========================
# CELL 14: CREATE DURATION IN MINUTES
# =========================
# duration_ms är svårt att tolka.
# duration_min är lättare att förstå i presentation och EDA.

df_clean_base["duration_min"] = df_clean_base["duration_ms"] / 60000

df_clean_base[["duration_ms", "duration_min"]].head()
```

```python
# =========================
# CELL 15: CLEANED DATASET WITH DUPLICATES
# =========================
# Detta dataset är renat men behåller track_id-dubbletter.
# Det är användbart om vi vill analysera samma låt i flera genre-sammanhang.

df_cleaned_with_duplicates = df_clean_base.copy()

print("Cleaned WITH duplicates shape:", df_cleaned_with_duplicates.shape)
print("Duplicate track_id:", df_cleaned_with_duplicates["track_id"].duplicated().sum())
```

```python
# =========================
# CELL 16: CLEANED DATASET WITH UNIQUE TRACKS
# =========================
# Detta dataset behåller bara en rad per track_id.
# Det passar bäst för klustring och unsupervised learning,
# eftersom samma låt annars kan påverka modellen flera gånger.

df_cleaned_unique_tracks = df_clean_base.drop_duplicates(
    subset=["track_id"],
    keep="first"
).copy()

print("Cleaned UNIQUE tracks shape:", df_cleaned_unique_tracks.shape)
print("Duplicate track_id:", df_cleaned_unique_tracks["track_id"].duplicated().sum())
```

```python
# =========================
# CELL 17: COMPARE THE TWO CLEANED DATASETS
# =========================
# Här gör vi en enkel jämförelse så teamet förstår skillnaden.

comparison = pd.DataFrame({
    "Dataset": [
        "spotify_cleaned_with_duplicates",
        "spotify_cleaned_unique_tracks"
    ],
    "Rows": [
        df_cleaned_with_duplicates.shape[0],
        df_cleaned_unique_tracks.shape[0]
    ],
    "Columns": [
        df_cleaned_with_duplicates.shape[1],
        df_cleaned_unique_tracks.shape[1]
    ],
    "Duplicate track_id": [
        df_cleaned_with_duplicates["track_id"].duplicated().sum(),
        df_cleaned_unique_tracks["track_id"].duplicated().sum()
    ],
    "Best used for": [
        "Genre analysis / same song in several genres",
        "Clustering / one row per song"
    ]
})

comparison
```

```python
# =========================
# CELL 18: SAVE CLEANED DATASETS
# =========================
# Vi sparar båda versionerna i data-mappen så att hela teamet kan använda samma filer.

with_duplicates_path = project_folder / "data" / "spotify_cleaned_with_duplicates.csv"
unique_tracks_path = project_folder / "data" / "spotify_cleaned_unique_tracks.csv"

df_cleaned_with_duplicates.to_csv(with_duplicates_path, index=False)
df_cleaned_unique_tracks.to_csv(unique_tracks_path, index=False)

print("Saved:", with_duplicates_path)
print("Exists:", with_duplicates_path.exists())

print("\nSaved:", unique_tracks_path)
print("Exists:", unique_tracks_path.exists())
```
